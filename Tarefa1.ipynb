**Análise Exploratória dos Dados**

Realize uma análise inicial dos dados:

○ Verifique a distribuição das variáveis (idade, renda, tempo no site, etc.).

○ Explore possíveis relações entre as variáveis independentes e a variável alvo (Compra).

○ Identifique valores ausentes ou inconsistências nos dados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arquivo = '/content/Dataset.csv'
```

```python
df = pd.read_csv(arquivo)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Idade'], kde=True, bins=30)
plt.title('Distribuição de Idade')
plt.show()

plt.figure(figsize=(8, 4))
sns.histplot(df['Renda Anual (em $)'], kde=True, bins=30, color='green')
plt.title('Distribuição da Renda Anual (em $)	')
plt.show()

plt.figure(figsize=(8, 4))
sns.histplot(df['Tempo no Site (min)'], kde=True, bins=30, color='red')
plt.title('Distribuição do Tempo no Site (min)')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gênero', color='pink')
plt.title('Distribuição de Gênero')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Anúncio Clicado', color='purple')
plt.title('Distribuição de Anúncio Clicado')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Compra (0 ou 1)', color='yellow')
plt.title('Distribuição de Compras')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Compra (0 ou 1)', y='Idade')
plt.title('Relaçao entre Idade e Compra')
plt.show()

plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Compra (0 ou 1)', y='Renda Anual (em $)')
plt.title('Relaçao entre Renda Anual e Compra')
plt.show()

plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Compra (0 ou 1)', y='Tempo no Site (min)')
plt.title('Relaçao entre Tempo no Site e Compra')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gênero', hue='Compra (0 ou 1)')
plt.title('Relação entre Gênero e Compra')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Anúncio Clicado', hue='Compra (0 ou 1)')
plt.title('Relação entre Anúncio Clicado e Compra')
plt.show()
```
